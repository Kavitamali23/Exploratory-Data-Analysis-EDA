```python
ohe=OneHotEncoder()
```

```python
##1. Use fit_transform method on categorical feature. Then convert it into a 2d array.
```

```python
degree_array=ohe.fit_transform(df[['degree_t']]).toarray()
```

```python
degree_array
```

```python
#Displays the categories of both variables
```

```python
degree_cat=ohe.categories_
degree_cat
```

```python
#Save both labels in a single array using ravel method
```

```python
degree_cat_arr=np.array(degree_cat).ravel()
degree_cat_arr
```

```python
df_degree=pd.DataFrame(degree_array,columns=degree_cat_arr)
df_degree
```

```python
df=pd.concat([df,df_degree],axis=1)
df
```

```python
df.drop('degree_t',axis=1,inplace=True)
df
```

```python
hsc_array=ohe.fit_transform(df[['hsc_s']]).toarray()
hsc_array
```

```python
hsc_cat=ohe.categories_
hsc_cat
```

```python
hsc_cat_arr=np.array(hsc_cat).ravel()
hsc_cat_arr
```

```python
df_hsc=pd.DataFrame(hsc_array,columns=hsc_cat_arr)
df_hsc
```

```python
df=pd.concat([df,df_hsc],axis=1)
df
```

```python
df.drop('hsc_s',axis=1,inplace=True)
```
