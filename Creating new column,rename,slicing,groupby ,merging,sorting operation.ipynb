```python
"""1.1 Create a new column 'Bonus'. Award 20% bonus of their salary for 'Sci&Tech' candidates.
1.2 In the ‘Bonus’ column created in 1.1, award 10% bonus of their salary for 'Comm&Mgmt' candidates.
1.3 In the ‘Bonus’ column created in 1.1, award 10% bonus of their salary for 'Others' candidates.
1.4 Rename columns as given below:-
Old Column Name New Column Name
sl_no SL_No.
ssc_p SSC %
hsc_p HSC %
degree_p Degree %
mba_p MBA %
"""
```
