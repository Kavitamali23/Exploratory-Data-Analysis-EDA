```python
## 1.IQR Method Capping and Trimming
```

```python
q1=df1['total_sqft_int'].quantile(0.25)
q3=df1['total_sqft_int'].quantile(0.75)
iqr=q3-q1
ul=q3+1.5*iqr
ll=q1-1.5*iqr
print("Q1:",q1)
print("Q3:",q3)
print("IQR:",iqr)
print("UL:",ul)
print("LL:",ll)
```

```python
## Trimming'
```

```python
df1.loc[(df1['total_sqft_int']>ul) | (df1['total_sqft_int']<ll)]
```

```python
df2=df1.loc[(df1['total_sqft_int']<ul) & (df1['total_sqft_int']>ll)]
df2
```

```python
sns.boxplot(df2['total_sqft_int'])
```

```python
# Capping
df3=df1.copy()
```

```python
df3.loc[df3['total_sqft_int']<ll,'total_sqft_int']=ll
df3.loc[df3['total_sqft_int']>ul,'total_sqft_int']=ul
```
