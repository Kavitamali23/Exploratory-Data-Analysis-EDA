### 6. Time Resampling using the resample() method

#### 6.1 For all the years what is the minimum, maximum value with the year end frequency ('A') option.

```python
# Resample the data by year-end frequency ('A') and calculate the min and max values
tsla_yearly_min = tsla.resample('A').min()
tsla_yearly_max = tsla.resample('A').max()

print("Yearly Minimum Values:")
print(tsla_yearly_min)

print("\nYearly Maximum Values:")
print(tsla_yearly_max)
```

#### 6.2 For all the years what is the minimum, maximum value with the Month Begin frequency ('MS') option.

```python
# Resample the data by month-start frequency ('MS') and calculate the min and max values 
tsla_monthly_min = tsla.resample('MS').min()
tsla_monthly_max = tsla.resample('MS').max()

print("Monthly Minimum Values (Month Begin):")
print(tsla_monthly_min)

print("\nMonthly Maximum Values (Month Begin):")
print(tsla_monthly_max)
```

#### 6.3 For all the years what is the minimum, maximum value with the Month End frequency ('M') option.

```python
# Resample the data by month-end frequency ('M') and calculate the min and max values
tsla_monthly_end_min = tsla.resample('M').min()
tsla_monthly_end_max = tsla.resample('M').max()

print("Monthly Minimum Values (Month End):")
print(tsla_monthly_min)

print("\nMonthly Maximum Values (Month End):")
print(tsla_monthly_max)
```

#### 6.4 For all the years what is the minimum, maximum value with the Quarter frequency ('QS') option.

```python
# Resample the data by quarter-start frequency ('QS') and calculate the min and max values 
tsla_quarterly_min = tsla.resample('QS').min()
tsla_quarterly_max = tsla.resample('QS').max()

print("Quarterly Minimum Values (Quarter Start):")
print(tsla_quarterly_min)

print("\nQuarterly Maximum Values (Quarter Start):")
print(tsla_quarterly_max)
```

#### 7.2 Plot for all the years max value with respect to Open using Month Begin frequency.

```python
plt.plot(tsla_monthly_max['Open'],color='blue')
plt.title('Monthly Maximum Values (Month Begin)')
plt.xlabel('Month')
plt.ylabel('Maximun Open price')
plt.legend()
plt.show()
```

#### 7.3 Plot for all the years max value with respect to Open using Month End frequency.

```python
plt.plot(tsla_monthly_end_max['Open'],color='purple')
plt.title('Monthly Maximum Values (Month End)')
plt.xlabel('Month')
plt.ylabel('Maximun Open price')
plt.legend()
plt.show()
```

#### 7.4 Plot for all the years max value with respect to Open using Quarter frequency.

```python
plt.plot(tsla_quarterly_max['Open'],color='green')
plt.title('Quarterly Maximum Values (Quarter Start)')
plt.xlabel('Quarterly')
plt.ylabel('Quarterly Maximum Values (Quarter Start)')
plt.xlabel('Quarter')
plt.ylabel('Maximun Open price')
plt.legend()
plt.show()
```
